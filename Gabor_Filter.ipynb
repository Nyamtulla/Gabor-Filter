```python
import matplotlib.pyplot as plt
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
pic = cv2.imread("/content/drive/MyDrive/DataScience/Gabor_Filter/pic.jpg")
print(pic.shape)
pic_resize = cv2.resize(pic, (400,400))
cv2_imshow(pic_resize)
```

```python
# cv2.getGaborKernel(ksize, sigma, theta, lambda, gamma, psi, ktype)
# ksize - size of gabor filter (n, n)
# sigma - standard deviation of the gaussian function
# theta - orientation of the normal to the parallel stripes
# lambda - wavelength of the sunusoidal factor
# gamma - spatial aspect ratio
# psi - phase offset
# ktype - type and range of values that each pixel in the gabor kernel can hold
def get_kernal(ksize,sigma,theta, lmd, gamma, psi):

  kernel = cv2.getGaborKernel(ksize, sigma, theta, lmd, gamma, psi, cv2.CV_32F)
  return kernel

kernel = get_kernal(ksize = (30,25),sigma=3,theta=50, lmd=8, gamma=0.8, psi=0)
g_kernel = cv2.resize(kernel, (300, 300), interpolation=cv2.INTER_CUBIC)
plt.imshow(kernel)
```

```python
img = cv2.cvtColor(pic_resize, cv2.COLOR_BGR2GRAY)
# cv2_imshow(img)
kernel1 = get_kernal(ksize = (20,20),sigma=3,theta=50, lmd=10, gamma=0.8, psi=0)
filtered_img = cv2.filter2D(img, cv2.CV_8UC3, kernel1)

cv2_imshow(filtered_img)
```

```python
fp = cv2.imread("/content/drive/MyDrive/DataScience/Gabor_Filter/fp.jpg")
#fp = cv2.resize(fp,(300,300))
plt.figure(figsize=(10,10))
plt.grid(which = 'both')
plt.imshow(fp)
```

```python
fp_gray = cv2.cvtColor(fp, cv2.COLOR_BGR2GRAY)
# cv2_imshow(img)
kernel2 = get_kernal(ksize = (5,5),sigma=2.7,theta=np.pi/6, lmd=10, gamma=0.7, psi=0)
kernel3 = get_kernal(ksize = (20,20),sigma=2.7,theta=2*np.pi/6, lmd=10, gamma=0.7, psi=0)
kernel4 = get_kernal(ksize = (20,20),sigma=2.7,theta=3*np.pi/6, lmd=10, gamma=0.7, psi=0)
kernel5 = get_kernal(ksize = (20,20),sigma=2.8,theta=4*np.pi/6, lmd=10, gamma=0.8, psi=0)
kernel6 = get_kernal(ksize = (20,20),sigma=3,theta=5*np.pi/6, lmd=10, gamma=0.8, psi=0)
kernel7 = get_kernal(ksize = (20,20),sigma=3,theta=6*np.pi/6, lmd=10, gamma=0.8, psi=0)
fig,ax = plt.subplots(3,2)
ax[0][0].imshow(kernel2)
ax[0][1].imshow(kernel3)
ax[1][0].imshow(kernel4)
ax[1][1].imshow(kernel5)
ax[2][0].imshow(kernel6)
ax[2][1].imshow(kernel7)

filtered = cv2.filter2D(fp_gray, cv2.CV_8UC3, kernel3)
filtered1 = cv2.filter2D(fp_gray, cv2.CV_8UC3, kernel5)
fig1,axe = plt.subplots(1,2)
axe[0].imshow(filtered)
axe[1].imshow(filtered1)
```
