```python
import matplotlib.pyplot as plt
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
pic = cv2.imread("/content/pexels-steve-johnson-985341.jpg")
print(pic.shape)
pic_resize = cv2.resize(pic, (400,400))
cv2_imshow(pic_resize)
```

```python
# cv2.getGaborKernel(ksize, sigma, theta, lambda, gamma, psi, ktype)
# ksize - size of gabor filter (n, n)
# sigma - standard deviation of the gaussian function
# theta - orientation of the normal to the parallel stripes
# lambda - wavelength of the sunusoidal factor
# gamma - spatial aspect ratio
# psi - phase offset
# ktype - type and range of values that each pixel in the gabor kernel can hold
def get_kernal(ksize,sigma,theta, lmd, gamma, psi):

  kernel = cv2.getGaborKernel(ksize, sigma, theta, lmd, gamma, psi, cv2.CV_32F)
  return kernel

kernel = get_kernal(ksize = (30,25),sigma=3,theta=50, lmd=8, gamma=0.8, psi=0)
g_kernel = cv2.resize(kernel, (300, 300), interpolation=cv2.INTER_CUBIC)
plt.imshow(kernel)
```

```python
img = cv2.cvtColor(pic_resize, cv2.COLOR_BGR2GRAY)
# cv2_imshow(img)
kernel1 = get_kernal(ksize = (20,20),sigma=3,theta=50, lmd=10, gamma=0.8, psi=0)
filtered_img = cv2.filter2D(img, cv2.CV_8UC3, kernel1)

cv2_imshow(filtered_img)
```
